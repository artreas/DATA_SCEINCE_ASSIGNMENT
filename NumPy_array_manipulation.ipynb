**Creating Arrays**

NumPy arrays are created using np.array().
image.png

```python
import numpy as np

array = np.array(np.random.randint(10,45,(8,2)))
print(array)
print(array.shape)
```

**Transpose**

Transpose swaps rows and columns.

```python
transpose_array = np.transpose(array)

print(transpose_array)
print(transpose_array.shape)
```

**3-D Arrays**

Create a 3D array like this:

```python
array = np.array(np.random.randint(10,45,(3,2,4)))
```

**Transpose of 3D Array**

```python
transposed_3d = np.transpose(array)
print(transposed_3d.shape)
```

**Reshape**

Reshape changes the shape but not the data.

```python
new_shape = (2,-1,6)

reshaped_array = np.reshape(array,new_shape)
```

**Resize**

Resize changes both shape and size.

```python
np.resize(array,(2,3,6))
```

**Flatten**

Flatten converts multi-dimensional array → 1D array.

```python
array.flatten()
```

**Insert**

Insert adds elements at a specific position.

```python
np.insert(array,1,100,axis=0)
```

**Append**

Append adds elements at the end.

```python


arr = np.array([[1,2,3],
                [4,5,6]])

new_row = [[7,8,9]]

result = np.append(arr, new_row, axis=0)

print(result)
arr = np.array([[1,2],
                [3,4]])

new_col = [[5],
           [6]]

result = np.append(arr, new_col, axis=1)

print(result)
```

Ceating new arrays

For basic numpy operations

```python


# Create two arrays
a = np.array([1,2,3,4])
b = np.array([5,6,7,8])

print("Array a:", a)
print("Array b:", b)
```

**Addition**

Adds two arrays element-wise.

```python
result = np.add(a,b)

print("Addition:", result)
```

**Subtraction**

Subtracts elements of one array from another.

```python
result = np.subtract(a,b)

print("Subtraction:", result)
```

**Multiplication**

Multiplies arrays element-wise.

```python
result = np.multiply(a,b)

print("Multiplication:", result)
```

**Division**

Divides elements of arrays.

```python
result = np.divide(b,a)

print("Division:", result)
```

**Sum of Elements**

Returns total sum of array elements.

```python
total = np.sum(a)

print("Sum:", total)
```

**Minimum Value**

Finds the smallest element.

```python
minimum = np.min(a)

print("Minimum value:", minimum)
```

**Maximum Value**

Finds the largest element.

```python
maximum = np.max(a)

print("Maximum value:", maximum)
```

**Mean (Average)**

Calculates average of elements.

```python
average = np.mean(a)

print("Mean:", average)
```

**Sorting an Array**

Sorts elements in ascending order.

```python
c = np.array([8,2,5,1])

sorted_array = np.sort(c)

print("Sorted array:", sorted_array)
```

**Broadcasting**

Broadcasting allows operations between arrays of different shapes.

```python
a = np.array([1,2,3,4])

result = a + 10

print("Broadcasting result:", result)
```
