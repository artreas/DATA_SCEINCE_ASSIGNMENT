**Convert columns if needed**

```python
df.columns
df['budget'] = pd.to_numeric(df['budget'], errors='coerce')
df['revenue'] = pd.to_numeric(df['revenue'], errors='coerce')
```

**Create Profit & ROI**

```python
df['profit'] = df['revenue'] - df['budget']
df['roi'] = df['profit'] / df['budget']
```

**Highest Profit Movie**

```python
max_profit_movie = df.loc[df['profit'].idxmax()]

max_profit_movie[['title', 'profit']]
import matplotlib.pyplot as plt

top_movies = df.sort_values(by='profit', ascending=False).head(10)

plt.figure()
plt.barh(top_movies['title'], top_movies['profit'])
plt.xlabel("Profit")
plt.ylabel("Movie")
plt.title("Top 10 Most Profitable Movies")
plt.gca().invert_yaxis()
plt.show()
```

**Language with Highest Avg ROI**

```python
# Cap extreme ROI values (95th percentile)
roi_cap = df['roi'].quantile(0.95)
df_clean = df[df['roi'] <= roi_cap]

# Compute average ROI
lang_roi = (
    df_clean.groupby('original_language')['roi']
    .mean()
    .sort_values(ascending=False)
    .head(10)
)

# Plot
plt.figure(figsize=(8,5))
plt.bar(lang_roi.index, lang_roi.values)

plt.xlabel("Language")
plt.ylabel("Average ROI")
plt.title("Top Languages by ROI (Outliers Removed)")
plt.xticks(rotation=45)

# Add labels
for i, v in enumerate(lang_roi.values):
    plt.text(i, v, f"{v:.2f}", ha='center', fontsize=9)

plt.tight_layout()
plt.show()
```

```python
import ast

# Extract actors
def get_actors(cast):
    try:
        cast_list = ast.literal_eval(cast)
        return [actor['name'] for actor in cast_list]
    except:
        return []

df['actors'] = df['cast'].apply(get_actors)

# Explode actors
df_actors = df.explode('actors')

# Top actor
top_actor = df_actors['actors'].value_counts().idxmax()
print("Top Actor:", top_actor)

# Filter + copy (important)
actor_movies = df_actors[df_actors['actors'] == top_actor].copy()

# Clean genres
def clean_genres(g):
    try:
        genre_list = ast.literal_eval(g)
        return ", ".join([item['name'] for item in genre_list])
    except:
        return ""

actor_movies['genres'] = actor_movies['genres'].apply(clean_genres)

# Sort by ROI (descending 🔥)
actor_movies = actor_movies.sort_values(by='roi', ascending=False)

# Reset index
actor_movies = actor_movies.reset_index(drop=True)

# Format values
actor_movies['profit'] = actor_movies['profit'].apply(lambda x: f"{x:,.0f}")
actor_movies['roi'] = actor_movies['roi'].round(2)

# Final output
actor_movies[['title', 'genres', 'profit', 'roi']]
```
