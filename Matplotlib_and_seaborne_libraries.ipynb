```python
# Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

**Load the Dataset**

```python
# Define dataset path
# Read the CSV file
df = pd.read_csv("company_sales_data.csv")

# View first rows
df.head()
# View first rows
df.head()
```

**Check Dataset Shape**

```python
# Check number of rows and columns
df.shape
```

Histogram **Plot**

```python
# =========================================
# Histogram of Total Profit
# =========================================

profit = df["total_profit"]

plt.hist(profit, bins=6, color="blue", edgecolor="black")

plt.xlabel("Total Profit")
plt.ylabel("Frequency")
plt.title("Distribution of Total Profit")

plt.show()
```

**Line Plot – Monthly Profit Trend**

```python
# Extract data
months = df["month_number"]
profit = df["total_profit"]

# Plot
plt.plot(months, profit, marker='o', color='blue')

# Labels
plt.xlabel("Month Number")
plt.ylabel("Total Profit")
plt.title("Monthly Profit Trend")

# Show plot
plt.show()
```

**Multi-Line Plot – Sales of All Products**

```python
months = df["month_number"]

plt.plot(months, df["facecream"], label="Facecream")
plt.plot(months, df["facewash"], label="Facewash")
plt.plot(months, df["toothpaste"], label="Toothpaste")
plt.plot(months, df["bathingsoap"], label="Bathing Soap")
plt.plot(months, df["shampoo"], label="Shampoo")
plt.plot(months, df["moisturizer"], label="Moisturizer")

plt.xlabel("Month Number")
plt.ylabel("Units Sold")
plt.title("Product Sales Data")

plt.legend()

plt.show()
```

**Bar Chart – Bathing Soap Sales**

```python

months = df["month_number"]
soap_sales = df["bathingsoap"]

plt.bar(months, soap_sales, color="purple")

plt.xlabel("Month Number")
plt.ylabel("Bathing Soap Sales")
plt.title("Monthly Bathing Soap Sales")

plt.show()
```

**Pie Chart – Product Sales Distribution**

```python
# Calculate total sales of each product
products = [
    df["facecream"].sum(),
    df["facewash"].sum(),
    df["toothpaste"].sum(),
    df["bathingsoap"].sum(),
    df["shampoo"].sum(),
    df["moisturizer"].sum()
]

labels = ["Facecream","Facewash","Toothpaste","Bathing Soap","Shampoo","Moisturizer"]

plt.pie(products, labels=labels, autopct="%1.1f%%")

plt.title("Sales Distribution by Product")

plt.show()
```

**Subplots – Compare Two Products**

```python
months = df["month_number"]

plt.subplot(2,1,1)
plt.plot(months, df["facecream"], color="blue")
plt.title("Facecream Sales")

plt.subplot(2,1,2)
plt.plot(months, df["facewash"], color="green")
plt.title("Facewash Sales")

plt.tight_layout()

plt.show()
```

**Scatter Plot – Profit vs Units**

```python

plt.scatter(df["total_units"], df["total_profit"], color="red")

plt.xlabel("Total Units Sold")
plt.ylabel("Total Profit")

plt.title("Units Sold vs Profit")

plt.show()
```

**Correlation Matrix (Prepare Data for Heatmap)**

```python

sales_data = df[[
    "facecream",
    "facewash",
    "toothpaste",
    "bathingsoap",
    "shampoo",
    "moisturizer",
    "total_units",
    "total_profit"
]]

# Compute correlation matrix
corr_df = sales_data.corr()

# Display correlation values
corr_df
```

Heatmap (Seaborn)

```python
plt.figure(figsize=(10,6))

sns.heatmap(corr_df,
            annot=True,
            cmap="coolwarm",
            linewidths=0.5)

plt.title("Correlation Heatmap of Product Sales")

plt.show()
```

**Seaborn Distribution Plot (Product Sales)**

```python
sns.displot(df["bathingsoap"], kde=True)

plt.title("Distribution of Bathing Soap Sales")

plt.show()
```

**Seaborn Joint Plot**

```python
sns.jointplot(
    x="bathingsoap",
    y="total_profit",
    data=df,
    kind="scatter"
)

plt.show()
```

**Seaborn Pair Plot**

```python
sns.pairplot(
    df[[
        "facecream",
        "facewash",
        "toothpaste",
        "bathingsoap",
        "shampoo",
        "moisturizer"
    ]],
    height=1.5   # reduce subplot size
)

plt.show()
```

**Box Plot (Outlier Detection)**

```python

plt.figure(figsize=(6,4))   # smaller plot size

sns.boxplot(
    data=df[[
        "facecream",
        "facewash",
        "toothpaste",
        "bathingsoap",
        "shampoo",
        "moisturizer"
    ]]
)

plt.title("Product Sales Distribution")
plt.xticks(rotation=45)

plt.show()
```
