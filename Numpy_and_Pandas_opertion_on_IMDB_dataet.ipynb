```python
# ================================
# Import Required Libraries
# ================================
import pandas as pd
import numpy as np


# ================================
# Load the Dataset
# ================================
# Read the IMDB dataset CSV file
df = pd.read_csv("imdb_data.csv")


# ================================
# Basic Dataset Exploration
# ================================

# Check the shape of the dataset (rows, columns)
print("Dataset Shape:", df.shape)

# Display first 5 rows of the dataset
print("\nFirst 5 Rows")
print(df.head())

# Display last 10 rows
print("\nLast 10 Rows")
print(df.tail(10))


# ================================
# Dataset Information
# ================================

# Display column names, datatypes and non-null values
print("\nDataset Info")
print(df.info())


# ================================
# Statistical Summary
# ================================

# Show statistical summary of numerical columns
print("\nStatistical Summary")
print(df.describe())


# ================================
# View All Column Names
# ================================

# Print all column names
print("\nColumns in Dataset")
print(df.columns)


# ================================
# Indexing and Slicing using iloc
# ================================

# iloc selects rows/columns using position indexes
# Here we select first 8 rows and first 4 columns
subset_iloc = df.iloc[:8, :4]

print("\nSubset using iloc (first 8 rows, first 4 columns)")
print(subset_iloc)


# ================================
# Indexing using loc
# ================================

# loc selects rows/columns using labels
# Here we select rows 50–60 and specific columns
subset_loc = df.loc[50:60, ['budget', 'genres', 'runtime']]

print("\nSubset using loc")
print(subset_loc)


# ================================
# Selecting Specific Columns
# ================================

# Extract selected columns from the dataframe
selected_columns = df[['original_language', 'release_date', 'cast']]

print("\nSelected Columns")
print(selected_columns.head())


# ================================
# Boolean Indexing
# ================================

# Check which rows have English as original language
english_condition = df['original_language'] == 'en'

print("\nBoolean Condition for English Movies")
print(english_condition.head())


# ================================
# Filtering English Movies
# ================================

# Create dataframe containing only English movies
english_movies = df[df['original_language'] == 'en']

print("\nEnglish Movies Dataset Shape")
print(english_movies.shape)


# ================================
# Multiple Condition Filtering
# ================================

# Filter movies that are English AND runtime > 120 minutes
long_english_movies = df[
    (df['original_language'] == 'en') &
    (df['runtime'] > 120)
]

print("\nLong English Movies Shape")
print(long_english_movies.shape)


# ================================
# Display Sample of Filtered Data
# ================================

print("\nSample of Long English Movies")
print(long_english_movies.head())
```

```python
# ==========================================================
# Pandas Column Manipulation, Operations and Data Cleaning
# ==========================================================

# ----------------------------------------------------------
# 1. Create New Columns
# ----------------------------------------------------------

# Create a new column that contains half of the movie runtime
df['half_runtime'] = df['runtime'] / 2

# Create a new column representing movie profit
# profit = revenue - budget
df['movie_profit'] = df['revenue'] - df['budget']

# View updated dataframe
df.head()



# ----------------------------------------------------------
# 2. Dropping Columns
# ----------------------------------------------------------

# Remove the column 'id'
df.drop('id', axis=1, inplace=True)

# Remove multiple columns at once
df.drop(['half_runtime','poster_path'], axis=1, inplace=True)

# Remove another column
df.drop(['crew'], axis=1, inplace=True)

# Preview dataset after dropping columns
df.head()



# ----------------------------------------------------------
# 3. Basic Column Operations
# ----------------------------------------------------------

# Sum of all movie budgets
total_budget = df['budget'].sum()
print("Total Budget:", total_budget)

# Sum of all runtimes
total_runtime = df['runtime'].sum()
print("Total Runtime:", total_runtime)

# Find minimum and maximum runtime
print("Shortest movie length:", df['runtime'].min())
print("Longest movie length:", df['runtime'].max())

# Find movies with runtime equal to zero
df[df['runtime'] == 0.0]



# ----------------------------------------------------------
# 4. Using apply() Method
# ----------------------------------------------------------

# Function to convert text to uppercase
def to_uppercase(column):
    return column.upper()

# Apply function to title column
df['Capitalized_title'] = df['title'].apply(to_uppercase)

df.head()



# ----------------------------------------------------------
# 5. Using Lambda Functions
# ----------------------------------------------------------

# Example lambda function
add_two_numbers = lambda a,b : a+b
print("Lambda Example:", add_two_numbers(1,5))


# Row-wise profit calculation using lambda
def revised_profit(revenue, budget):

    if revenue > 0:
        new_profit = revenue - budget
    else:
        new_profit = np.nan

    return new_profit


# Apply function row-wise
df['new_profit'] = df.apply(lambda x: revised_profit(x['revenue'], x['budget']), axis=1)

df.head()



# ----------------------------------------------------------
# 6. Sorting Data
# ----------------------------------------------------------

# Sort movies by runtime in descending order
sorted_movies = df.sort_values('runtime', ascending=False)

sorted_movies.head()



# ----------------------------------------------------------
# 7. Datetime Operations
# ----------------------------------------------------------

# Convert release_date column to datetime format
df['release_date'] = pd.to_datetime(df['release_date'], errors='coerce')

# Extract year from release date
df['release_year'] = df['release_date'].dt.year

# Extract month
df['release_month'] = df['release_date'].dt.month

# Extract day
df['release_day'] = df['release_date'].dt.day

df.head()



# ----------------------------------------------------------
# 8. Exploring Unique Values
# ----------------------------------------------------------

# Find unique languages in dataset
unique_languages = df['original_language'].unique()
print("Unique Languages:", unique_languages)

# Count number of unique languages
print("Number of Unique Languages:", df['original_language'].nunique())

# Count occurrences of each language
print("Language Counts:\n", df['original_language'].value_counts())



# ----------------------------------------------------------
# 9. Handling Missing Values
# ----------------------------------------------------------

# Check missing values in each column
print("Missing Values in Each Column:\n")
print(df.isnull().sum())

# Fill missing values with a placeholder
df_filled = df.fillna("Unknown")

# Preview dataset after filling missing values
df_filled.head()
```
